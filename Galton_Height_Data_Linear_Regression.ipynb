```python
slope = (((df['Father']-df['Father'].mean())*(df['Height']-df['Height'].mean())).sum()) / ((df['Father']-df['Father'].mean())**2).sum()
```

```python
slope
```

```python
intercept = y.mean() - slope * X.mean() 
```

```python
intercept
```
